```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown
```

```python
load_dotenv(override=True)
openrouter_api_key = os.getenv("OPENROUTER_API_KEY")

if openrouter_api_key:
    print(f"OpenRouter API key loaded: {openrouter_api_key[:8]}")



```

```python
openai = OpenAI(api_key=openrouter_api_key, base_url="https://openrouter.ai/api/v1")
```

```python
messages = [{"role": "user", "content": "What is 2+2?"}]
```

```python
response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=messages,
)
print(response.choices[0].message.content)
```

```python
question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=messages,
)

generated_question = response.choices[0].message.content

print(generated_question)
```

```python
question = [{'role': 'user', 'content': generated_question}]

response = openai.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=question
)

answer = response.choices[0].message.content

display(Markdown(f"**Question:** {generated_question}\n\n**Answer:** {answer}"))




```
